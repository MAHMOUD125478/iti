### 1.B Slice the first two columns using `.loc` and store the result in `result_1`
شرح: هنا هنستخدم أسماء الأعمدة عبر `df.columns[0:2]` و`.loc` علشان نرجع كل الصفوف والأول عمودين.

```python
try:
    result_1 = df.loc[:, df.columns[0:2]]
    print("result_1 shape:", result_1.shape)
    display(result_1.head())
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```

### 1.C Slice the first two rows using `.loc` and store the result in `result_2`
شرح: في `.loc` التقطيع شامل للحد الأخير، فنستخدم `0:1` علشان نجيب الصفوف 0 و1.

```python
try:
    result_2 = df.loc[0:1, :]
    print("result_2 shape:", result_2.shape)
    display(result_2)
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```

### 1.D Slice the first four rows and the first five columns and store in `result_3`
هنا نستخدم `.iloc` لأننا نعتمد على الفهرس الرقمي.

```python
try:
    result_3 = df.iloc[0:4, 0:5]
    print("result_3 shape:", result_3.shape)
    display(result_3)
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```

### 1.E Slice rows 0,4,6 and columns invoice, time and price into `result_4`
هنا نستخدم `.loc` مع لستة الصفوف وأسماء الأعمدة.

```python
try:
    cols_needed = ["invoice", "time", "price"]
    missing_cols = [c for c in cols_needed if c not in df.columns]
    if missing_cols:
        print("الأعمدة التالية مش موجودة في الداتا:", missing_cols)
        print("الأعمدة المتاحة:", list(df.columns[:20]))
    else:
        result_4 = df.loc[[0,4,6], cols_needed]
        print("result_4 shape:", result_4.shape)
        display(result_4)
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```

### 1.F Store the number of rows in `num_rows` and print the last row
سنحسب عدد الصفوف وبعدين نطبع آخر صف باستخدام `iloc`.

```python
try:
    num_rows = df.shape[0]
    print("Number of rows (num_rows):", num_rows)
    last_row = df.iloc[num_rows - 1]
    print("\nLast row values:")
    display(last_row)
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```

## Part 2 - مثال على حساب المتوسط والحد الأقصى (avg_price, max_price)
الأسئلة: Compute the average and max price charged. Store in `avg_price` و `max_price`.

```python
try:
    price_col = None
    for candidate in ["price", "Price", "unit_price", "UnitPrice", "Amount", "amount"]:
        if candidate in df.columns:
            price_col = candidate
            break
    if price_col is None:
        print("لم أجد عمود سعر واضح. الأعمدة المتاحة:", list(df.columns[:30]))
    else:
        avg_price = df[price_col].mean()
        max_price = df[price_col].max()
        print(f"Using column: {price_col}")
        print("avg_price =", avg_price)
        print("max_price =", max_price)
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```

### مثال آخر: حساب متوسط السعر ومتوسط عدد العناصر المشتراة
لو عندنا عمود `num_items` أو `Quantity` هنحسب المتوسطين ونخزنهم في `avg_price` و`avg_num_items`. إذا مش موجودين سيعرض رسالة توضيحية.

```python
try:
    qty_candidates = ["num_items", "NumItems", "quantity", "Quantity", "items", "Items"]
    qty_col = next((c for c in qty_candidates if c in df.columns), None)
    if price_col is None:
        print("لا يوجد عمود سعر معروف لاستعماله في avg_price هنا.")
    else:
        avg_price = df[price_col].mean()
    if qty_col is None:
        print("لم أجد عمود للـ quantity. الأعمدة المتاحة:", list(df.columns[:30]))
    else:
        avg_num_items = df[qty_col].mean()
        print("avg_price =", avg_price)
        print("avg_num_items =", avg_num_items)
except NameError:
    print("DataFrame 'df' غير موجود. أولًا شغّل خلية قراءة الملف.")
```
